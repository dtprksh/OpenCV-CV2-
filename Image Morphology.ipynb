```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img =cv2.imread('shyam.jpg',0)
```

```python
ret,thres=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(thres)
```

### Eroding of image

```python
kernel=np.ones((9,9),np.uint8)
erosion=cv2.erode(img,kernel,iterations=5)
plt.imshow(erosion)
```

### Dilating the image

```python
dilation=cv2.dilate(img,kernel,iterations=5)
```

```python
plt.imshow(dilation)
```

### Closing(erosion followed by dilation)

```python
dilero=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
plt.imshow(dilero)
```

### Opening(Dilation followed by erosion)

```python
erodil=cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.imshow(erodil)
```

### Difference Between dilation and erosion

```python
gradient=cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
plt.imshow(gradient)
```

### difference between input image and Opening of the image.

```python
tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kernel)
plt.imshow(tophat)
```

### difference between the closing of the input image and input image.

```python
blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kernel)
plt.imshow(blackhat)
```
