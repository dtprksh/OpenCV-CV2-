```python
import numpy as np
import cv2
```

```python
img=cv2.imread('landscape.jpeg')
```

#### Reading pixel value of a coordinate

```python
px= img[10,100]
print(px)
```

#### finding blue and green pixel of a certain coordinate

```python
blue=img[156,78,0]
green=img[67,89,1]
print(blue)
print(green)
```

#### changing pixel value of a coordinate

```python
img[79,32]=[255,0,255]          # not a good method to access pixel
print(img[79,32])
```

#### better way of accessing pixel

```python
img.item(56,74,2)
```

#### changing red pixel value of a certain of a certain coordinate

```python
img.itemset((56,74,2),200)
img.item(56,74,2)
```

#### array shape of image

```python
img.shape
```

#### image datatype

```python
img.dtype    # Note img.dtype is very important while debugging because a large number of errors
            # in OpenCV-Python code is caused by invalid datatype.
```

#### putting the tree at bottom in the middle of sky

```python
tree=img[200:325,40:140]
img[10:135,300:400]=tree
cv2.imshow('landscape',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### splits the image into different color aspects (costly operation in terms of time)

```python
b,g,r=cv2.split(img)
cv2.imshow('landscape',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### make all red pixel in image to zero

```python
img[:,:,2]=0
cv2.imshow('landscape',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### make all green pixels zero

```python
img[:,:,1]=0
cv2.imshow('landscape',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### make all blue pixels = 40

```python
img[:,:,0]=40
cv2.imshow('landscape',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### adds different types of padding in image

```python
import matplotlib.pyplot as plt
%matplotlib inline
green=[0,255,0]
plt.figure(figsize=(40,30))
constant=cv2.copyMakeBorder(img,18,18,18,18,cv2.BORDER_CONSTANT,value=green)
reflect=cv2.copyMakeBorder(img,19,19,19,19,cv2.BORDER_REFLECT)
wrap=cv2.copyMakeBorder(img,20,20,20,20,cv2.BORDER_WRAP)
replicate=cv2.copyMakeBorder(img,21,21,21,21,cv2.BORDER_REPLICATE)
default=cv2.copyMakeBorder(img,22,22,22,22,cv2.BORDER_DEFAULT)
plt.subplot(231),plt.imshow(img),plt.title('ORIGINAL')
plt.xticks([]),plt.yticks([])
plt.subplot(232),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.xticks([]),plt.yticks([])
plt.subplot(233),plt.imshow(reflect),plt.title('REFLECT')
plt.xticks([]),plt.yticks([])
plt.subplot(234),plt.imshow(wrap),plt.title('WRAP')
plt.xticks([]),plt.yticks([])
plt.subplot(235),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.xticks([]),plt.yticks([])
plt.xticks([]),plt.yticks([])
plt.subplot(236),plt.imshow(default,'gray'),plt.title('DEFAULT')
plt.xticks([]),plt.yticks([])
```
