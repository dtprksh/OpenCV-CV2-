```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('shyam.jpg',0)
plt.imshow(img)
```

# Edge Detection

### Laplacian Filter

```python
laplace=cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(laplace)
```

```python
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=31)
plt.imshow(sobelx)
```

```python
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=31)
plt.imshow(sobely)
```

```python
sobelx8u=cv2.Sobel(img,cv2.CV_8U,1,0,ksize=5)
sobelx64f=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
abs_soblex=np.absolute(sobelx64f)
sobel_8u=np.uint8(abs_soblex)
```

```python
plt.imshow(sobel_8u)
```

```python
plt.imshow(sobelx8u)
```

### canny edge detection

```python
edge=cv2.Canny(img,5,38)
plt.imshow(edge)
```
