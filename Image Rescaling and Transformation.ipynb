```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

#### BGR to RGB due to matplotlib and cv2

```python
img=cv2.imread('shyam.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

#### Resizing with different types of interpolation(inter-cubic for slow, inter-linear for zoom, inter-area for shrinking

```python
res=cv2.resize(img,None,fx=1,fy=3,interpolation=cv2.INTER_CUBIC)
plt.imshow(res)
```

```python
res1=cv2.resize(img,None,fx=5,fy=1,interpolation=cv2.INTER_LINEAR)
plt.imshow(res1)
```

```python
res2=cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
plt.imshow(res2)
```

```python
warp=cv2.imread('ansuj.jpg',0)
plt.imshow(warp)
```

#### translating image along x and y axis

```python
rows,cols=warp.shape
M=np.float32([[1,0,1000],[0,1,500]])
dst=cv2.warpAffine(warp,M,(cols,rows))
plt.imshow(dst)
```

#### Rotating an image by an angle

```python
cht=cv2.imread('shyam.jpg',0)
rows1,cols1=cht.shape
M=cv2.getRotationMatrix2D((cols1/2,rows1/2),60,1)
dst1=cv2.warpAffine(cht,M,(cols1,rows1))
plt.imshow(dst1)
```

```python
M=cv2.getRotationMatrix2D((cols1/2,rows1/2),-120,1)
dst2=cv2.warpAffine(dst1,M,(cols1,rows1))
plt.imshow(dst2)
```

#### Affine Transformation (Transforming the original image by taking three point and respectively moving it to three different points
#### causing the image to transform)

```python
km=cv2.imread('km.jpg',0)
rows,cols=km.shape
pts1=np.float32([[50,200],[300,400],[1000,2000]])
pts2=np.float32([[150,300],[450,600],[800,1700]])
M1=cv2.getAffineTransform(pts1,pts2)
dst3=cv2.warpAffine(km,M1,(cols,rows))
plt.imshow(dst3)
```

#### Selecting four points on image and making them 4 end points for the new image causing image to transform

```python
pts3=np.float32([[1300,200],[1300,2600],[2700,200],[2600,2700]])
pts4=np.float32([[0,0],[0,4000],[3100,0],[3100,4000]])
M2=cv2.getPerspectiveTransform(pts3,pts4)
dst4=cv2.warpPerspective(km,M2,(3100,4000))
plt.imshow(dst4)
```
