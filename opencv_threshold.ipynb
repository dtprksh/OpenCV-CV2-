```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('ansuj.jpg',0)
```

```python
ret,thres1=cv2.threshold(img,120,255,cv2.THRESH_BINARY)
ret,thres2=cv2.threshold(img,120,255,cv2.THRESH_BINARY_INV)
ret,thres3=cv2.threshold(img,120,255,cv2.THRESH_TRUNC)
ret,thres4=cv2.threshold(img,120,255,cv2.THRESH_TOZERO)
ret,thres5=cv2.threshold(img,120,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thres1, thres2, thres3, thres4, thres5]
```
