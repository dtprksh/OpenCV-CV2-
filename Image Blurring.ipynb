```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('ansuj.jpg')
```

### Converting RGB to BGR

```python
img=cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

### Averaging the filters using filter2D

```python
kernel =np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img,-1,kernel)
plt.imshow(dst)
```

### Blurring using Box Filter (All filter size below should be odd and positive)

```python
blur=cv2.blur(img,(55,55))
plt.imshow(blur)
```

### Gaussian Blurring (Gaussian Kernel is used,needs std dev in x and y direction if y is not given assumes it smae as x)

```python
gaussian_blur=cv2.GaussianBlur(img,(51,51),17)
plt.imshow(gaussian_blur)
```

### Finds the median of area and puts it in central box

```python
median_blur=cv2.medianBlur(img,61)
plt.imshow(median_blur)
```

### Applies Gaussian filtering but doest not removes edges

```python
bilateral_filter=cv2.bilateralFilter(img,9,75,75)
plt.imshow(bilateral_filter)
```
